```python
#Ouverture des datas
data = pd.read_csv('data_litt.txt',sep=",",index_col=False)
print(data)
```

```python
data = data.sort_values(by=['HJD'])
```

```python
data.to_csv('data-litterature.txt', index=False, header=True, sep=',', float_format='%.4f')
```

```python
date = data['HJD'] + 2443000
mag = data['V [mag]']
```

```python
mid = 2443878.253
P = 2.592891
phase = np.mod((date-mid)/P,1)
```

```python
fig, ax = plt.subplots()

ax.plot(phase, mag, lw=1, color='k', marker='s', linestyle = 'none', label=('Data litérature'))
#ax.invert_yaxis()
#ax.set_xlim(-0.05,0.95)


ax.set_title('Courbe de luminosité de AX Vel', pad=20)
ax.set_ylabel('Magnitude V [mag]')
ax.set_xlabel('Phase')
ax.legend(loc='best', frameon=True)

#plt.savefig('graphes/litterature.pdf', bbox_inches='tight')
plt.show()
```
